# 01 Data Understanding

This notebook loads and inspects the PCOS and endometriosis datasets. It prints shapes, columns, dtypes, missing values, duplicates, and class balance for the target columns. It also identifies the exact target column names for both datasets.

```python
# Imports
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Path setup

PROJECT_ROOT = Path.cwd().parent
PCOS_PATH = PROJECT_ROOT / "data" / "raw" / "pcos" / "PCOS_data_without_infertility.xlsx"
ENDO_PATH = PROJECT_ROOT / "data" / "raw" / "endometriosis" / "structured_endometriosis_data.csv"
```

```python
# Load PCOS Excel file
df_pcos = pd.read_excel(PCOS_PATH)
df_pcos.head()

xls = pd.ExcelFile(PCOS_PATH)
xls.sheet_names

```

```python

df_pcos = pd.read_excel(PCOS_PATH, sheet_name="Full_new")
df_pcos.head()
df_pcos.shape
df_pcos.columns
```

```python
# Load Endometriosis CSV file
df_endo = pd.read_csv(ENDO_PATH)
df_endo.head(20)
```

```python
# PCOS: shape
df_pcos.shape
```

```python
# Endometriosis: shape
df_endo.shape
```

```python
# PCOS: columns
df_pcos.columns.tolist()
```

```python
# Endometriosis: columns
df_endo.columns.tolist()
```

```python
# Endometriosis: dtypes
df_endo.dtypes
```

```python
# PCOS: missing values
df_pcos.isna().sum()
```

```python
# Endometriosis: missing values
df_endo.isna().sum()
```

```python
# PCOS: duplicated rows
df_pcos.duplicated().sum()
```

```python
# Endometriosis: duplicated rows
df_endo.duplicated().sum()
```

```python
# PCOS: class balance (target column)
# Try common target column names
for col in ['PCOS (Y/N)', 'PCOS', 'target', 'label']:
    if col in df_pcos.columns:
        print(f'PCOS target column: {col}')
        print(df_pcos[col].value_counts(dropna=False))
        break
else:
    print('No obvious PCOS target column found!')
```

```python
# Endometriosis: class balance (target column)
# Try common target column names
for col in ['Diagnosis', 'target', 'label', 'Endometriosis', 'Endo']:
    if col in df_endo.columns:
        print(f'Endometriosis target column: {col}')
        print(df_endo[col].value_counts(dropna=False))
        break
else:
    print('No obvious endometriosis target column found!')
```

```python
# Save simple feature summary tables
meta_dir = Path('..') / 'data' / 'metadata'
meta_dir.mkdir(exist_ok=True, parents=True)
df_pcos.describe(include='all').T.to_csv(meta_dir / 'pcos_feature_summary.csv')
df_endo.describe(include='all').T.to_csv(meta_dir / 'endo_feature_summary.csv')
```

```python
import os
def relpath(p):
    return os.fspath(Path(p).relative_to(PROJECT_ROOT)) if Path(p).is_absolute() else str(p)

# Show PCOS feature summary
pcos_summary_path = meta_dir / 'pcos_feature_summary.csv'
endo_summary_path = meta_dir / 'endo_feature_summary.csv'
print(f"Saved PCOS feature summary to {relpath(pcos_summary_path)}")
print(f"Saved endometriosis feature summary to {relpath(endo_summary_path)}")
pd.read_csv(pcos_summary_path).head(10)
pd.read_csv(endo_summary_path).head(10)
```
